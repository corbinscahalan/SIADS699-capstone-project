```python
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
```

```python
image_folder = 'nn_images/cnn_images'
results_path = 'cnn_test_results.json.gz'
data_path = 'cnn_data_test.json.gz'
```

```python
results = pd.read_json(results_path)
test_data = pd.read_json(data_path)
```

```python
results['residuals'] = results['log_vid_viewcount'] - results['cnn_thumb_preds']
```

```python
sorted_results = results.sort_values(by='residuals')
```
